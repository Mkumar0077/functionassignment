Q1.  **What is the difference between a function and a method in Python?**
  - A **function** in Python is a block of reusable code that performs a specific task. It can be called independently, without being attached to any object.

```python
# Example:-
def greet(name):
      return f"Hello, {name}!"

print(greet("Mr.Kumar"))
```

  - A **method**, on the other hand, is a function that is defined inside a class and is associated with objects of that class. Methods are called on an instance of the class, and they can access and modify the object's attributes.

```python
class Person:
      def __init__(self, name):
          self.name = name

      def greet(self):
          return f"Hello, {self.name}!"

person = Person("Mr.Kumar")
print(person.greet())
```

Q5.**What are iterators in Python and how do they differ from iterables?**
 - An **iterator** is an object in Python that implements two special methods: __iter__() and __next__(). It allows sequential access to items in a collection one at a time, until there are no more items to retrieve.

- An **iterable** is any object in Python that can return an iterator using the iter() function. Examples of iterables include lists, strings, tuples, and dictionaries

***Key Differences:-***
1. Iterables can be converted into iterators using iter(), whereas iterators are already ready for iteration.
2. Iterators maintain state and can only be traversed once. Iterables can be used to generate fresh iterators again and again.

```python
#Example:-
# Iterable
my_list = [1, 2, 3]

# Converting iterable to iterator
my_iterator = iter(my_list)

# Accessing elements using an iterator
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

Q6.**Explain the concept of generators in Python and how they are defined.**
 - A **generator** is a special type of iterator in Python that allows you to yield values one at a time using the yield keyword, rather than returning all values at once like a regular function.
 - Generators are useful for handling large datasets or infinite sequences because they don't store all the values in memory—they generate each value when needed.

***How to define a generator:-***
Using a function with the yield keyword.

```python
#example:-
def my_generator():
    yield 1
    yield 2
    yield 3

# Calling the generator
gen = my_generator()
print(next(gen))
print(next(gen))
print(next(gen))
```

Q7.**What are the advantages of using generators over regular functions?**
 - 1. ***Memory Efficiency:-*** Generators produce values one at a time using the yield keyword. This avoids storing all the values in memory at once, which is particularly useful for processing large datasets or infinite sequences.
   
   2. ***Lazy Evaluation:-*** Generators generate values only when needed, allowing for efficient computation and avoiding unnecessary overhead.
   
   3. ***Improved Performance:-*** Since values are computed on demand, generators can be faster than regular functions for sequences where not all values are needed.

   4. ***Easier Implementation:-*** Generators simplify code for iterating over large sequences compared to manually creating iterators.


```python
#example:-
# Generator for even numbers
def even_numbers(limit):
    num = 0
    while num <= limit:
        yield num
        num += 2

# Using the generator
for even in even_numbers(10):
    print(even) #Here, the generator produces even numbers up to the specified limit one at a time.
```

Q8. **What is a lambda function in Python and when is it typically used?**
 - - A **lambda function** is an anonymous function in Python. It doesn’t have a name and is defined using the lambda keyword.
- It can have any number of arguments but only a single expression. The result of the expression is automatically returned.

- Lambda functions are typically used for:
1. *Short-term use cases:* When a simple function is needed temporarily.
2. *Inline operations:* Within functions like map(), filter(), or reduce().
3. *Concise code:* To simplify operations without the need for formal def syntax.

```python
#Example:-
# Using lambda to double numbers in a list
numbers = [1, 2, 3, 4]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)
```

Q9. **Explain the purpose and usage of the `map()` function in Python.**
 - *Purpose:*
The map() function applies a given function to every item in an iterable (e.g., list, tuple), and returns a map object containing the results.

*Usage:*
1. It’s used for *transforming data* in iterables efficiently.
2. Especially useful for *concise and functional programming.*


```python
#example:-
# Function to convert Celsius to Fahrenheit
def to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
temperatures = [0, 20, 30]

# Applying map to transform temperatures
fahrenheit = list(map(to_fahrenheit, temperatures))
print(fahrenheit)
```

Q10. **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

 - **map() function:-** Applies a given function to every item in an iterable and returns a map object containing the results.

```python
#example:-
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

 - **reduce() function:-** Reduces an iterable to a single value by applying a function cumulatively to its items. (Requires the functools module.)

```python
#example:-
from functools import reduce
numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)
```

 - ***filter() function:-*** *italicized text* Filters items in an iterable based on a given condition, returning only the items that meet the condition.

```python
numbers = [1, 2, 3, 4]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

- ***Key Differences:-***
1. map() transforms items.
2. reduce() combines items into a single value.
3. filter() selects items based on a condition.

```python
'''Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.'''
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(numbers))
```

```python
'''Q4.Write a Python function that checks if a given number is prime or not from 1 to 200.'''
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):  # Check divisors up to √n
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

```python
'''Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.'''
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
exponent = 5
for power in power_of_two(exponent):
    print(power)
```

```python
'''Q7. Implement a generator function that reads a file line by line and yields each line as a string.'''
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
'''Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.'''
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
temperatures_celsius = [0, 20, 30, 40]

# Using map to apply the conversion function
temperatures_fahrenheit = list(map(celsius_to_fahrenheit, temperatures_celsius))

print("Temperatures in Fahrenheit:", temperatures_fahrenheit)
```

```python
'''Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
'''
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, text))

# Example usage:
text = "Hello, World!"
result = remove_vowels(text)
print(result)
```

Q11.) **Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:**

- order no.---- Book title and Author ---------- Quantity--------- Price per item
- 34587-------- learning python, Mark Lutz ----- 4  ------------------- 40.95
- 98762-------- Programing Python,Mark Lutz----- 5  ------------------- 56.80
- 77226-------- Head first python,Poul Barry ----- 3 ------------------ 32.95
- 88112-------- Einfuhrung in python3,Brend Klein---3 ----------------- 24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

**Write a Python program using lambda and map.**

```python
# Given list with sublists
orders = [
    [34587, "learning python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Poul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Brend Klein", 3, 24.99]
]

# Lambda function to calculate the final amount
calculate_total = lambda order: (order[0], order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3])

# Using map to apply the lambda function to the orders list
result = list(map(calculate_total, orders))

# Print the resulting list with 2-tuples
print(result)
```
